# **Mount Google Drive & Load Dataset**

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

path='/content/drive/My Drive/Week 4_ANNS JANA/Week 4.xlsx'
df=pd.read_excel(path)

df.head()
```

# **Basic Exploration**

```python
print("Rows, Columns:",df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())
```

# **Data Cleaning**

```python
# Remove duplicate rows
df=df.drop_duplicates()

# Remove extra spaces from column names
df.columns=df.columns.str.strip()

# Check missing values again
print(df.isnull().sum())

# Convert Date column to datetime
df['Date']=pd.to_datetime(df['Date'])

# Verify
df.info()
```

# **Feature Engineering**

```python
df['Profit Margin %']=(df['Profit']/df['Sales'])*100

df[['Sales','Profit','Profit Margin %']].head()
```

# **Findings**

##Top Countries by Sales

```python
country_sales=df.groupby('Country')['Sales'].sum().sort_values(ascending=False)

print(country_sales.head(10))
```

##Top Products by Profit

```python
product_profit=df.groupby('Product')['Profit'].sum().sort_values(ascending=False)

print(product_profit.head(10))
```

##Sales by Segment

```python
segment_sales=df.groupby('Segment')['Sales'].sum()

print(segment_sales)
```

##Monthly Sales Trend

```python
monthly_sales=df.groupby('Month Name')['Sales'].sum().sort_values(ascending=False)

print(monthly_sales)
```

##Country-wise Profit

```python
country_profit=df.groupby('Country')['Profit'].sum().sort_values(ascending=False)

print(country_profit)
```

##Discount Impact on Profit

```python
discount_profit=df.groupby('Discount Band')['Profit'].mean()

print(discount_profit)
```

# **Download Cleaned Dataset**

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from google.colab import files

# Load file
path='/content/drive/My Drive/Week 4_ANNS JANA/Week 4.xlsx'
df=pd.read_excel(path)

# Cleaning
df=df.drop_duplicates()
df.columns=df.columns.str.strip()

# Fill missing values
for c in df.select_dtypes(include='number'):
    df[c]=df[c].fillna(df[c].median())

for c in df.select_dtypes(include='object'):
    df[c]=df[c].fillna(df[c].mode()[0])

cleaned_file='Week4_Cleaned.xlsx'
df.to_excel(cleaned_file,index=False)

files.download(cleaned_file)
```
